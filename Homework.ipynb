```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Problem 2.1
#https://see.stanford.edu/materials/lsoeldsee263/homeworkProblems.pdf

G=np.array([[1,0.2,0.1],[0.1,2,0.1],[0.3,0.1,3]])

gamma=3
alpha=1.2
sigma=0.01
```

```python
p=np.array([[1],[3],[7]])
s=np.diag(np.diag(G))@p
q=sigma+(G-np.diag(np.diag(G)))@p
q
```

```python
ts=np.array([[0],[0],[0]])
ts2=np.array([[0],[0],[0]])
for t in range(0,25):
    S=(np.diag(np.diag(G))@p)/(sigma+(G-np.diag(np.diag(G)))@p)  
    p=alpha*gamma*p/S
    ts2=np.hstack((ts2,S))
    ts=np.hstack((ts,p))
    
```
